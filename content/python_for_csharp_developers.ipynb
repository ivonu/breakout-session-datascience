## 1. Basic Syntax Comparison

### Variables and Type Declaration

````python
# C# Style:
# int accountNumber = 123456789;
# string customerName = "John Smith";
# double accountBalance = 15000.75;
# bool isActive = true;

# Python Style (Dynamic Typing):
account_number = 123456789
customer_name = "John Smith"
account_balance = 15000.75
is_active = True  # Note: True/False capitalized in Python

print(f"Account Number: {account_number} (type: {type(account_number).__name__})")
print(f"Customer Name: {customer_name} (type: {type(customer_name).__name__})")
print(f"Account Balance: ${account_balance} (type: {type(account_balance).__name__})")
print(f"Account Active: {is_active} (type: {type(is_active).__name__})")
````

````python
# Python Type Hints (Optional - Similar to C# typing)
from typing import List, Dict, Optional

# With type hints (Python 3.5+)
customer_id: int = 67890
customer_name: str = "Sarah Johnson"
account_types: List[str] = ["Checking", "Savings", "Investment"]
account_details: Dict[str, str] = {"type": "Premium", "tier": "Gold"}
assigned_banker: Optional[str] = None  # Similar to nullable types in C#

print(f"Customer: {customer_name} (ID: {customer_id})")
print(f"Available Account Types: {account_types}")
print(f"Account Details: {account_details}")
print(f"Assigned Banker: {assigned_banker}")

customer_id = "12345"  # No error at runtime, but type hint indicates this should be an int
print(f"Customer: {customer_name} (ID: {customer_id})")
````

### String Operations and Formatting

````python
# C# String Interpolation:
# string message = $"Hello {name}, you are {age} years old";

# Python String Formatting (Multiple Ways):
customer_name = "Alice Thompson"
account_balance = 2500.75

# 1. f-strings (Python 3.6+, most similar to C#)
message1 = f"Hello {customer_name}, your account balance is ${account_balance:.1f}"

# 2. .format() method
message2 = "Hello {}, your account balance is ${:.1f}".format(customer_name, account_balance)
message3 = "Hello {name}, your account balance is ${balance:.1f}".format(name=customer_name, balance=account_balance)

# 3. % formatting (older style)
message4 = "Hello %s, your account balance is $%.1f" % (customer_name, account_balance)

print("f-string:        ", message1)
print(".format():       ", message2)
print("named .format(): ", message3)
print("% formatting:    ", message4)
````

## 2. Control Structures

### Conditional Statements

````python
# C# Style:
# if (creditScore >= 750) {
#     rating = "Excellent";
# } else if (creditScore >= 700) {
#     rating = "Good";
# } else {
#     rating = "Fair";
# }

# Python Style (Note: No braces, indentation matters!)
credit_score = 720

if credit_score >= 750:
    rating = "Excellent"
elif credit_score >= 700:  # Note: elif, not "else if"
    rating = "Good"
else:
    rating = "Fair"

print(f"Credit Score: {credit_score}, Rating: {rating}")

# Python Ternary Operator (similar to C# ternary)
# C#: string status = isActive ? "Active" : "Inactive";
account_active = True
status = "Active" if account_active else "Dormant"
print(f"Account Status: {status}")
````

### Loops

````python
# For loops comparison

# C# Style:
# for (int i = 0; i < 5; i++) {
#     Console.WriteLine(i);
# }

# Python Style:
print("Processing account IDs:")
for i in range(5):  # range(5) generates 0, 1, 2, 3, 4
    account_id = f"ACC{1000 + i}"
    print(account_id)

print("\nProcessing transaction amounts:")
for amount in range(100, 1000, 200):  # start=100, stop=1000, step=200
    print(f"${amount}")

# Iterating over collections (more Pythonic)
account_types = ["Checking", "Savings", "Credit"]

print("\nAvailable account types:")
for account_type in account_types:
    print(account_type)

print("\nAccount types with indices:")
for index, account_type in enumerate(account_types):
    print(f"{index + 1}: {account_type}")
````

````python
# While loops
# Similar to C#, but no braces

transaction_count = 0
while transaction_count < 3:
    print(f"Processing transaction #{transaction_count + 1}")
    transaction_count += 1  # Note: No ++ operator in Python

# Python has no do-while, but you can simulate it
print("\nSimulated do-while - checking account balance:")
balance_check = 0
while True:
    print(f"Balance check #{balance_check + 1}: Account verified")
    balance_check += 1
    if balance_check >= 3:
        break
````

### Exception Handling

````python
# C# Style:
# try {
#     int result = int.Parse(userInput);
#     Console.WriteLine(result);
# } catch (FormatException ex) {
#     Console.WriteLine("Invalid format");
# } catch (Exception ex) {
#     Console.WriteLine($"Error: {ex.Message}");
# } finally {
#     Console.WriteLine("Cleanup");
# }

# Python Style:
account_input = "abc12345"  # This will cause an error when parsing as int

try:
    account_number = int(account_input)
    print(f"Valid account number: {account_number}")
except ValueError as ex:  # Similar to FormatException in C#
    print(f"Invalid account number format: {ex}")
except Exception as ex:   # Catch-all exception
    print(f"Error processing account: {ex}")
else:  # Python-specific: runs if no exception occurred
    print("Account number successfully validated!")
finally:
    print("Account validation process completed")
````

## 3. Functions and Methods

### Function Definition and Parameters

````python
# C# Method:
# public static double CalculateInterest(double principal, double rate) {
#     return principal * rate;
# }

# Python Function:
def calculate_interest(principal, rate):
    return principal * rate

# With type hints (optional)
def calculate_compound_interest(principal: float, rate: float, years: int) -> float:
    return principal * (1 + rate) ** years

result1 = calculate_interest(10000, 0.05)
result2 = calculate_compound_interest(10000.0, 0.05, 5)

print(f"Simple interest on $10,000 at 5%: ${result1:,.2f}")
print(f"Compound interest after 5 years: ${result2:,.2f}")
````

````python
# Default parameters and named arguments

# C# Style (method overloading or optional parameters):
# public static void GreetCustomer(string name, string greeting = "Welcome") {
#     Console.WriteLine($"{greeting} to First National Bank, {name}!");
# }

# Python Style:
def greet_customer(name, greeting="Welcome", bank_name="First National Bank"):
    return f"{greeting} to {bank_name}, {name}!"

# Different ways to call the function
print(greet_customer("Alice"))                           # Using defaults
print(greet_customer("Bob", "Hello"))                    # Positional arguments
print(greet_customer("Charlie", greeting="Good morning")) # Named argument
print(greet_customer("Diana", bank_name="Community Bank", greeting="Welcome")) # Named, different order
print(greet_customer(name="Eve", greeting="Thank you for choosing")) # All named
````

````python
# Variable arguments (*args and **kwargs)
# Similar to C# params keyword

# C#: public static double CalculateTotalBalance(params double[] balances)
# Python equivalent:
def calculate_total_balance(*balances):  # *args accepts any number of positional arguments
    return sum(balances)

def create_account_profile(**kwargs):  # **kwargs accepts any number of named arguments
    profile = {}
    for key, value in kwargs.items():
        profile[key] = value
    return profile

def process_transaction(*args, **kwargs):
    print(f"Transaction amounts: {args}")
    print(f"Transaction details: {kwargs}")

print("Total Balance:", calculate_total_balance(1500.50, 2300.75, 850.00, 4200.25))
print("Account Profile:", create_account_profile(type="Premium", tier="Gold", overdraft_protection=True))
print("\nProcessing transaction:")
process_transaction(250.00, 150.75, account="ACC001", description="ATM Withdrawal")
````

## 4. Object-Oriented Programming

### Classes and Objects

````python
# C# Class:
# public class BankAccount {
#     public string AccountHolder { get; set; }
#     public string AccountNumber { get; set; }
#
#     public BankAccount(string accountHolder, string accountNumber) {
#         AccountHolder = accountHolder;
#         AccountNumber = accountNumber;
#     }
#     
#     public string GetAccountInfo() {
#         return $"Account: {AccountNumber} - {AccountHolder}";
#     }
# }

# Python Class:
class BankAccount:
    def __init__(self, account_holder, account_number):  # Constructor (similar to C# constructor)
        self.account_holder = account_holder             # Public attribute (like C# public property)
        self.account_number = account_number
        self._account_type = "Standard"                  # "Protected" (by convention, single underscore)
        self.__balance = 0.0                            # "Private" (name mangling, double underscore)

    def get_account_info(self):                         # Method (note: self is explicit in Python)
        return f"Account: {self.account_number} - {self.account_holder}"

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            raise ValueError("Deposit amount must be positive")

    def get_balance(self):
        return self.__balance

# Create and use objects
account1 = BankAccount("Alice Johnson", "ACC001234")
account1.deposit(1500.75)

print(account1.get_account_info())
print(f"Balance: ${account1.get_balance():,.2f}")
print(f"Account Holder (direct access): {account1.account_holder}")
````

````python
# Python Properties (similar to C# properties)
class BetterBankAccount:
    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number
        self._balance = 0.0

    @property  # Getter (similar to C# get accessor)
    def balance(self):
        return self._balance

    @balance.setter  # Setter (similar to C# set accessor)
    def balance(self, value):
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = value

    @property
    def formatted_balance(self):  # Read-only property
        return f"${self._balance:,.2f}"

    def __str__(self):  # Similar to C# ToString() method
        return f"BankAccount({self.account_holder}, {self.account_number})"

    def __repr__(self):  # For debugging (more technical representation)
        return f"BetterBankAccount(account_holder='{self.account_holder}', account_number='{self.account_number}')"

account2 = BetterBankAccount("Bob Smith", "ACC005678")
account2.balance = 2500.50  # Using property setter

print(f"Account balance: {account2.balance}")              # Using property getter
print(f"Formatted balance: {account2.formatted_balance}")  # Read-only property
print(f"String representation: {account2}")               # __str__ method
print(f"Debug representation: {repr(account2)}")          # __repr__ method
````

### Inheritance

````python
# C# Inheritance:
# public class PremiumAccount : BankAccount {
#     public List<Transaction> Transactions { get; set; }
#
#     public PremiumAccount(string holder, string number) : base(holder, number) {
#         Transactions = new List<Transaction>();
#     }
#     
#     public override string GetAccountInfo() {
#         return base.GetAccountInfo() + $" - Premium with {Transactions.Count} transactions";
#     }
# }

# Python Inheritance:
class PremiumAccount(BetterBankAccount):  # Inherits from BetterBankAccount
    def __init__(self, account_holder, account_number, interest_rate=0.05):
        super().__init__(account_holder, account_number)  # Call parent constructor
        self.interest_rate = interest_rate
        self.transactions = []

    def add_transaction(self, amount, description):
        transaction = {"amount": amount, "description": description}
        self.transactions.append(transaction)

    def __str__(self):  # Override parent method
        base_info = super().__str__()
        return f"{base_info} - Premium with {len(self.transactions)} transactions"

    @property
    def monthly_interest(self):
        return self.balance * (self.interest_rate / 12)  # Monthly interest calculation

# Create premium account and add transactions
premium_account = PremiumAccount("Sarah Connor", "PREM001", 0.045)
premium_account.balance = 15000.00

premium_account.add_transaction(500.00, "Direct Deposit")
premium_account.add_transaction(-75.50, "Utility Payment")

print(premium_account)
print(f"Monthly interest: {premium_account.monthly_interest:,.2f}")
print(f"Total balance with interest: ${premium_account.balance + premium_account.monthly_interest:,.2f}")
````

### Interfaces and Abstract Classes

````python
# C# Interface and Abstract Class:
# public interface IAccount {
#     void Deposit(double amount);
#     void Withdraw(double amount);
#     double GetBalance();
# }
#
# public abstract class AbstractAccount : IAccount {
#     protected double balance;
#
#     public abstract void Deposit(double amount);
#     public abstract void Withdraw(double amount);
#
#     public virtual double GetBalance() {
#         return balance;
#     }
# }

# Python ABC (Abstract Base Class) - Similar to C# interfaces/abstract classes
from abc import ABC, abstractmethod

# Interface-like class (all methods are abstract)
class IAccount(ABC):
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def get_balance(self):
        pass

# Abstract class with some concrete and some abstract methods
class AbstractBankAccount(IAccount):
    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number
        self._balance = initial_balance

    # Concrete method (like virtual in C#)
    def get_balance(self):
        return self._balance

    def get_account_info(self):
        return f"Account {self.account_number}: ${self._balance:,.2f}"

    # Abstract methods - must be implemented by subclasses
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def calculate_fees(self):
        pass

# Concrete implementations
class StandardAccount(AbstractBankAccount):
    def __init__(self, account_number, initial_balance=0):
        super().__init__(account_number, initial_balance)
        self.transaction_fee = 2.50

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return f"Deposited ${amount:,.2f}. New balance: ${self._balance:,.2f}"
        raise ValueError("Deposit amount must be positive")

    def withdraw(self, amount):
        fee = self.calculate_fees()
        total_amount = amount + fee
        if total_amount <= self._balance:
            self._balance -= total_amount
            return f"Withdrew ${amount:,.2f} (Fee: ${fee:.2f}). New balance: ${self._balance:,.2f}"
        raise ValueError("Insufficient funds")

    def calculate_fees(self):
        return self.transaction_fee

class PremiumBankAccount(AbstractBankAccount):
    def __init__(self, account_number, initial_balance=0):
        super().__init__(account_number, initial_balance)
        self.minimum_balance = 1000

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return f"Premium Deposit: ${amount:,.2f}. New balance: ${self._balance:,.2f}"
        raise ValueError("Deposit amount must be positive")

    def withdraw(self, amount):
        fee = self.calculate_fees()
        total_amount = amount + fee
        if total_amount <= self._balance:
            self._balance -= total_amount
            return f"Premium Withdrawal: ${amount:,.2f} (Fee: ${fee:.2f}). New balance: ${self._balance:,.2f}"
        raise ValueError("Insufficient funds")

    def calculate_fees(self):
        # Premium accounts have no fees if balance is above minimum
        return 0.0 if self._balance >= self.minimum_balance else 5.0

# Usage examples
print("=== Abstract Classes and Interfaces in Python ===")

# Cannot instantiate abstract class
try:
    # This would raise TypeError: Can't instantiate abstract class
    abstract_account = AbstractBankAccount("ABS001")
except TypeError as e:
    print(f"Cannot instantiate abstract class: {e}")

# Create concrete implementations
standard = StandardAccount("STD001", 500)
premium = PremiumBankAccount("PREM001", 2000)

accounts = [standard, premium]

for account in accounts:
    print(f"\n{account.get_account_info()}")
    print(account.deposit(100))
    print(account.withdraw(50))
    print(f"Current fees: ${account.calculate_fees():.2f}")
````

````python
# Protocol (Python 3.8+) - More Pythonic alternative to interfaces
from typing import Protocol, runtime_checkable

# Protocol defines the interface without inheritance
@runtime_checkable
class Transactable(Protocol):
    def process_payment(self, amount: float) -> str:
        ...

    def get_account_id(self) -> str:
        ...

# Classes that implement the protocol (no explicit inheritance needed)
class CreditCard:
    def __init__(self, card_number, credit_limit=1000):
        self.card_number = card_number
        self.balance = 0
        self.credit_limit = credit_limit

    def process_payment(self, amount: float) -> str:
        if self.balance + amount <= self.credit_limit:
            self.balance += amount
            return f"Credit Card payment of ${amount:.2f} processed"
        return "Payment declined - credit limit exceeded"

    def get_account_id(self) -> str:
        return f"CC-{self.card_number}"

class DebitCard:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def process_payment(self, amount: float) -> str:
        if amount <= self.balance:
            self.balance -= amount
            return f"Debit Card payment of ${amount:.2f} processed"
        return "Payment declined - insufficient funds"

    def get_account_id(self) -> str:
        return f"DB-{self.account_number}"

class DigitalWallet:
    def __init__(self, wallet_id, balance=0):
        self.wallet_id = wallet_id
        self.balance = balance

    def process_payment(self, amount: float) -> str:
        if amount <= self.balance:
            self.balance -= amount
            return f"Digital Wallet payment of ${amount:.2f} processed"
        return "Payment declined - insufficient funds"

    def get_account_id(self) -> str:
        return f"DW-{self.wallet_id}"

# Function that works with any object implementing Transactable protocol
def process_payments(payment_methods: list[Transactable], amount: float):
    for method in payment_methods:
        print(f"{method.get_account_id()}: {method.process_payment(amount)}")

# Protocol usage - no explicit inheritance needed!
payment_methods = [
    CreditCard("1234-5678", 2000),
    DebitCard("ACC001", 500),
    DigitalWallet("WALLET001", 300)
]

print("\n=== Protocol-based Interface (Python 3.8+) ===")
process_payments(payment_methods, 100.0)

# Runtime type checking with protocols
print("\n=== Runtime Protocol Checking ===")
for method in payment_methods:
    is_transactable = isinstance(method, Transactable)
    print(f"{method.get_account_id()} implements Transactable: {is_transactable}")
````

## 5. Collections and Data Structures

### Lists (Similar to C# List<T>)

### Dictionaries (Similar to C# Dictionary<TKey, TValue>)

````python
# C# Dictionary:
# var customer = new Dictionary<string, object> {
#     {"name", "Alice Thompson"},
#     {"accountNumber", "ACC001234"},
#     {"balance", 2500.75}
# };
# customer["accountType"] = "Premium";
# bool hasKey = customer.ContainsKey("name");

# Python Dictionary:
customer = {
    "name": "Alice Thompson",
    "account_number": "ACC001234",
    "balance": 2500.75
}

customer["account_type"] = "Premium"     # Add new key-value pair
has_key = "name" in customer             # Check if key exists

print(f"Customer: {customer}")
print(f"Name: {customer['name']}")
print(f"Has 'name' key: {has_key}")

# Safe access (similar to TryGetValue)
balance = customer.get("balance", 0.0)           # Returns 0.0 if key doesn't exist
credit_limit = customer.get("credit_limit", "Not specified")
print(f"Balance: ${balance:,.2f}")
print(f"Credit Limit: {credit_limit}")

# Dictionary methods
print(f"Keys: {list(customer.keys())}")
print(f"Values: {list(customer.values())}")
print(f"Items: {list(customer.items())}")
````

````python
# Dictionary iteration
customer_credit_scores = {
    "Alice Thompson": 750,
    "Bob Johnson": 680,
    "Charlie Brown": 720,
    "Diana Wilson": 800
}

print("Customer Credit Scores:")
for name in customer_credit_scores:                    # Iterate over keys
    print(f"  {name}: {customer_credit_scores[name]}")

print("\nUsing items() for key-value pairs:")
for name, score in customer_credit_scores.items():     # Iterate over key-value pairs
    rating = "Excellent" if score >= 750 else "Good" if score >= 700 else "Fair"
    print(f"  {name}: {score} (Rating: {rating})")

# Dictionary comprehension (Python-specific)
excellent_customers = {name: score for name, score in customer_credit_scores.items() if score >= 750}
print(f"\nExcellent credit customers: {excellent_customers}")
````

## 6. Distinctive Python Features

### List Comprehensions (No direct C# equivalent)

````python
# C# LINQ equivalent:
# var squares = amounts.Select(x => x * x).ToList();
# var positiveAmounts = amounts.Where(x => x > 0).ToList();
# var positiveSquares = amounts.Where(x => x > 0).Select(x => x * x).ToList();

# Python List Comprehensions:
transaction_amounts = [100, -50, 200, -25, 150, -75, 300, -100, 250, -30]

# Basic comprehension
squares = [x * x for x in transaction_amounts]                      # Square all amounts
positive_amounts = [x for x in transaction_amounts if x > 0]        # Filter positive amounts
positive_squares = [x * x for x in transaction_amounts if x > 0]    # Filter and transform

print(f"Original amounts: {transaction_amounts}")
print(f"Squares: {squares}")
print(f"Positive amounts: {positive_amounts}")
print(f"Positive amount squares: {positive_squares}")

# More complex examples with banking terms
account_types = ["checking", "savings", "investment", "credit"]
type_lengths = [len(account_type) for account_type in account_types]
long_types = [account_type.upper() for account_type in account_types if len(account_type) > 6]

print(f"\nAccount types: {account_types}")
print(f"Type name lengths: {type_lengths}")
print(f"Long type names (uppercase): {long_types}")

# Nested comprehension for account matrix
account_matrix = [[1001, 1002, 1003], [2001, 2002, 2003], [3001, 3002, 3003]]
flattened_accounts = [account_id for row in account_matrix for account_id in row]
print(f"\nAccount matrix: {account_matrix}")
print(f"Flattened account IDs: {flattened_accounts}")
````

### Duck Typing (Very different from C# static typing)

````python
# "If it processes like a bank account and transacts like a bank account, it's a bank account"
# Python cares about behavior, not explicit type declarations

# In C#, you'd need interfaces or inheritance
# In Python, if objects have the same methods, they can be used interchangeably

class CheckingAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def process_transaction(self, amount):
        self.balance += amount
        return f"Checking Account {self.account_number}: ${amount:,.2f} processed. New balance: ${self.balance:,.2f}"

class SavingsAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def process_transaction(self, amount):
        self.balance += amount
        return f"Savings Account {self.account_number}: ${amount:,.2f} processed. New balance: ${self.balance:,.2f}"

class CreditCard:
    def __init__(self, card_number, credit_limit=1000):
        self.card_number = card_number
        self.balance = 0  # Credit cards start at 0
        self.credit_limit = credit_limit

    def process_transaction(self, amount):
        if self.balance + amount <= self.credit_limit:
            self.balance += amount
            return f"Credit Card {self.card_number}: ${amount:,.2f} charged. Balance: ${self.balance:,.2f}"
        else:
            return f"Credit Card {self.card_number}: Transaction declined - would exceed limit"

# Function that works with any object that has process_transaction() method
def process_batch_transactions(accounts, amount):
    for account in accounts:
        print(account.process_transaction(amount))  # Duck typing: we don't care about the type!

# Create different account objects
accounts = [
    CheckingAccount("CHK001", 1000),
    SavingsAccount("SAV001", 5000),
    CreditCard("CC001", 2000)
]

print("Processing $250 transaction on all accounts:")
process_batch_transactions(accounts, 250)  # Works with all objects that have process_transaction()
````

### Decorators (Somewhat similar to C# Attributes, but more powerful)

````python
# Decorators modify or enhance functions/methods
# Similar concept to C# attributes, but executed at runtime

import time
import functools

# Simple decorator
def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} took {end_time - start_time:.4f} seconds")
        return result
    return wrapper

# Decorator with parameters
def log_calls(prefix="CALL"):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            print(f"[{prefix}] Calling {func.__name__} with args: {args}, kwargs: {kwargs}")
            result = func(*args, **kwargs)
            print(f"[{prefix}] {func.__name__} returned: {result}")
            return result
        return wrapper
    return decorator

# Using decorators
@timer
@log_calls("MATH")
def calculate_sum(n):
    """Calculate sum of numbers from 1 to n"""
    return sum(range(1, n + 1))

# Function is automatically wrapped by decorators
result = calculate_sum(1000000)
print(f"Final result: {result}")
````
